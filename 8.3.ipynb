```python
def adfuller_test(p):
  result = adfuller(p)
  labels = ['test statisitc', 'p-value', '#lags', '#observation']
  for value, label in zip(result, labels):
    print(label +':' +str(value))
  if result[1] <= 0.05:
    print('stationary')
  else:
    print('non-stationary')
```

```python
import warnings
import itertools
import statsmodels.api as sm
warnings.filterwarnings("ignore")
p=d=q=range(0, 3)
pdq=list(itertools.product(p,d,q))
pdq_x_PDQs = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p,d,q))]
a=[]
b=[]
c=[]
wf=pd.DataFrame()
for param in pdq:
  for seasonal_param in pdq_x_PDQs:
    try:
      mod = sm.tsa.statespace.SARIMAX(pd_x['チケット'], order=param, seasonal_order=seasonal_param,enforce_invertibility=False, enforce_stationarity=False)
      results = mod.fit()
      print('ARIMA{}x{} - AIC:{}'.format(param, seasonal_param, results.aic))
      a.append(param)
      b.append(seasonal_param)
      c.append(results.aic)
    except:
      continue
wf['pdq'] = a
wf['pdq_x_PDQs'] = b
wf['aic'] = c
print(wf[wf['aic']==wf['aic'].min()])
```
