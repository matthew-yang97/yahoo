```python
from collections import defaultdict
order_dict = defaultdict(tuple)
import statsmodels.api as sm
#plt.figure(figsize=[15, 7])
#sm.tsa.seasonal_decompose(pd_x['off'][300:330]).plot()

import warnings
import itertools
import statsmodels.api as sm
warnings.filterwarnings("ignore")
print(all_list)
```

```python
for cols in all_list:
  col, r = cols
  if col in order_dict:
    continue
  p=d=q=range(0, 3)
  pdq=list(itertools.product(p,d,q))
  pdq_x_PDQs = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p,d,q))]
  a=[]
  b=[]
  c=[]
  wf=pd.DataFrame()
  for param in pdq:
    for seasonal_param in pdq_x_PDQs:
      try:
        mod = sm.tsa.statespace.SARIMAX(pd_x[col], order=param, seasonal_order=seasonal_param,enforce_invertibility=False, enforce_stationarity=False)
        results = mod.fit()
        print('ARIMA{}x{} - AIC:{}'.format(param, seasonal_param, results.aic))
        a.append(param)
        b.append(seasonal_param)
        c.append(results.aic)
      except:
        continue
  wf['pdq'] = a
  wf['pdq_x_PDQs'] = b
  wf['aic'] = c
  order_dict[col] = (wf[wf['aic']==wf['aic'].min()])
  print(order_dict[col])
```
