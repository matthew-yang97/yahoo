```
def emergency_process(p,l):
  # 
  # for i in range(l):
  #   if p[i]==0:
  #     p[i]=0
  #     continue
  #   if p[i]==1:
  #     count=0
  #   if p[i]==2 or p[i]==3:
  #     p[i]=p[i]+count
  #     count+=1
   for i in range(l):
    if p[i]==0:
      p[i]=-1
      continue
    if p[i]==1 or p[i]==2 or p[i]==3:
      p[i]=1

    return p
```

```
df_e = pd.read_csv("emergency.csv", parse_dates=[0], index_col=0)
for i in city_kanji:
  df_e[i]=emergency_process(df_e[i],len(df_e[i]))
df_e.describe()
```

check correlation for emergency

```
from scipy.stats import pearsonr
df_e.describe()
cols = np.array(df_e.columns)
list_plus = []
list_minus = []
for i in city_kanji:
  for col in cols:
    #adfuller_test(df_e[col])
    r, p = pearsonr(df_e[col], df[i])
    print(r, p)
    if r > 0.55:
      list_plus.append(col)
    if r < -0.55:
      list_minus.append(col)
  print(list_plus)
  list_plus.clear()
  print(list_minus)
  list_minus.clear()
  print(i)

```
